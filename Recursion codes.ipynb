```python
def factorial(n):
    """Assumes n is an integer >= 0, returns n!"""
    assert n>=0, f'{n} should be an integer greater than or equal to zero'
    if n < 1:
        return 1
    else:
        return n* factorial(n-1)
print(factorial(5))

#Factorial function executed through iteration rather than recursion
def iterFactorial(n):
    result=1
    while n>=1:
        result*=n
        n-=1
    return result
print(iterFactorial(5))
```

```python
#Recursion
#Find the largest number in a list using recursion
def maximum(arr):
    """Assumes maximum is a list containing numbers(floats or integers)
       Returns the largest number that occurs in this list"""
    if len(arr)==1:
        return arr[0]
    else:
        return arr[0] if arr[0]>=maximum(arr[1:]) else maximum(arr[1:])
print(maximum([2,7,9,6]))

```

```python
#Finding the sum of two or more numbers
def summation(arr):
    """Assumes arr is a list of numbers
       Returns the sum of all the elements within the list"""
    if len(arr)==1:
        return arr[0]
    else:
        return arr[0]+summation(arr[1:])
print(summation([10,9,8]))

```

```python
#Finding the product of two or more numbers
def product(arr):
    """Assumes arr is a list of numbers
       Returns the product of all the elements within the list"""
    if len(arr)==1:
        return arr[0]
    else:
        return arr[0]*product(arr[1:])
    
print(product([1]))
print(product([10,9]))
print(product([1,2,3,4]))
```

```python
#Implementation of the longest common subsequence in Python
def LCS(x,y):
    m,n=len(x),len(y)
    L=[[None]*(n+1)for i in range(m+1)]#filling an (m+1)(n+1)table with null values
    for i in range(m+1):
        for j in range(n+1):
            if i==0 or j==0:
                L[i][j]=0
            elif x[i-1]==y[j-1]:
                L[i][j]=L[i-1][j-1]+1
            else:
                L[i][j]=max(L[i-1][j],L[i][j-1])
    print('The length of the longest common subsequence is',L[m][n])
    #printing the actual LCS
    solution=[]
    while L[m][n]>0:
        if x[m-1]==y[n-1]:
            solution.append(x[m-1])
            m-=1
            n-=1
        elif L[m-1][n]>=L[m][n-1]:
            m-=1
        else:
            n-=1
    return ''.join(reversed(solution))
            
               
seq1=input("Enter the first sequence: ")
seq2=input("Enter the second sequence: ")
print(LCS(seq1,seq2))


                    
                


            
            
        
```

```python
from queue import PriorityQueue

class Graph:
    def __init__(self, vertex_count):
        self.vertices = vertex_count
        self.edges = [[-1 for i in range(vertex_count)] for j in range(vertex_count)]
        #edges will be an adjacency matrix(list of lists) describing weight of edges from a node A to a node B 
        #self.edges[i][j]=weight from node i to node j
        self.visited = [] #list keeping track of visited vertices
    def add_edge(self, u, v, weight): 
        """Adds nodes with their weight to the graph"""
        self.edges[u][v]=weight
        self.edges[v][u]=weight

def dijkstra(graph, source): #source represents the starting vertex in the graph
    D = {v:float('inf') for v in range(graph.vertices)} 
    #The dictionary above initializes all shortest paths from the source to be infinity
    D[source] = 0
    #shortest path from the source vertex to the source vertex is set to zero

    pq = PriorityQueue()#tracks vertices with updated shortest path after relaxation on the vertices
    pq.put((0, source))

    while not pq.empty():
        (dist, current_vertex) = pq.get()#gets vertices from the priorityqueue based on the least cost
        graph.visited.append(current_vertex) #vertex got from the priorityqueue is marked as visited

        for neighbor in range(graph.vertices):
            if graph.edges[current_vertex][neighbor] != -1:
                distance = graph.edges[current_vertex][neighbor]#getting the weight between current vertex and neighbor
                if neighbor not in graph.visited:
                    old_cost = D[neighbor]
                    new_cost = D[current_vertex] + distance #relaxation of vertices
                    if new_cost < old_cost: #comparing to get the minimum cost
                        pq.put((new_cost, neighbor))
                        D[neighbor] = new_cost
    return D
#Implementation of Dijkstra's algorithm with a sample graph

g = Graph(9) #graph will have nine vertices
#Edges describing weight from one node to the next
g.add_edge(0, 1, 4)
g.add_edge(0, 6, 7)
g.add_edge(1, 6, 11)
g.add_edge(1, 7, 20)
g.add_edge(1, 2, 9)
g.add_edge(2, 3, 6)
g.add_edge(2, 4, 2)
g.add_edge(3, 4, 10)
g.add_edge(3, 5, 5)
g.add_edge(4, 5, 15)
g.add_edge(4, 7, 1)
g.add_edge(4, 8, 5)
g.add_edge(5, 8, 12)
g.add_edge(6, 7, 1)
g.add_edge(7, 8, 3) 

D=dijkstra(g,0)
for vertex in range(len(D)):
    print("The shortest path from vertex 0 to vertex",vertex,"is",D[vertex])
    


        
```

```python

class Graph:
    def __init__(self, node_count):
        self.node_count = node_count
        self.edges = [[0 for i in range(node_count)] for j in range(node_count)]
       #edges will be an adjacency matrix(list of lists) describing weight of edges from a node A to a node B 
       # Initialized with  0
       #self.edges[i][j]=weight from node i to node j
    def add_edge(self, u, v, weight): 
        """Adds nodes with their weight to the graph"""
        self.edges[u][v]=weight
        self.edges[v][u]=weight
    def prims_mstree(self):
        selected_nodes = [False for node in range(self.node_count)]
        #This list initializes all nodes with a false value indicating they have not yet been added to the MS tree
        
        result = [[0 for column in range(self.node_count)]
                  for row in range(self.node_count)]#Adjacency matrix to store the MS tree
                    
        edge_no = 0 #counts the edges in the MS tree
        
        # While there are nodes that are not included in the MST, keep looking:
        while(False in selected_nodes):
         # For every vertex in the set S, find the all adjacent vertices
         #calculate the distance from the vertex selected at step 1.
         # if the vertex is already in the set S, discard it otherwise
         # choose another vertex nearest to selected vertex  at step 1.
            minimum = float('inf') #minimum is set to infinity so that a node with less weight can be selected as the starting node
            start = 0 #starting node
            end = 0 #ending node

            for i in range(self.node_count):
                # If the node is part of the MS tree, look at its relationships
                if selected_nodes[i]:
                    for j in range(self.node_count):
                        if (not selected_nodes[j] and self.edges[i][j]>0): #an edge exists and the node has not yet been selected 
                            if self.edges[i][j] < minimum:
                                # Defines the new minimum weight, the starting vertex and the ending vertex
                                minimum = self.edges[i][j]
                                start, end = i, j
        
            # Since we added the ending vertex to the MST, it's already selected:
            selected_nodes[end] = True

            # Filling the MS tree Adjacency Matrix fields:
            if minimum == float('inf'):
                result[start][end]=0 #there is a less costly edge including those nodes in the MS tree
            else:
                result[start][end]=minimum
                
            result[end][start] = result[start][end]

        # Print the resulting MS tree
        print("Edge : Weight")
        # for node1, node2, weight in result:
        for i in range(len(result)):
            for j in range(i, len(result)):
                if result[i][j] != 0:
                    edge_no+=1 #increment the number of edges by 1
                    print(str(i), "-",str(j),":",str(result[i][j]))
        print("Total number of edges in the tree is",edge_no)
#implementation on a sample graph
g = Graph(9) #graph will have nine nodes
#Edges describing weight from one node to the next
g.add_edge(0, 1, 4)
g.add_edge(0, 6, 7)
g.add_edge(1, 6, 11)
g.add_edge(1, 7, 20)
g.add_edge(1, 2, 9)
g.add_edge(2, 3, 6)
g.add_edge(2, 4, 2)
g.add_edge(3, 4, 10)
g.add_edge(3, 5, 5)
g.add_edge(4, 5, 15)
g.add_edge(4, 7, 1)
g.add_edge(4, 8, 5)
g.add_edge(5, 8, 12)
g.add_edge(6, 7, 1)
g.add_edge(7, 8, 3)

print(g.prims_mstree())
```
